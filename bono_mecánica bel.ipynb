```python
import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(0, 80, 10000)
w1 = 5
inicial = 0.01

def calcularfase(t):
  phi= w1 * t
  return phi

fase = calcularfase(t)


def calcularresistencia(phase):
  re = (inicial / 2) * np.exp(fase) + (inicial / 2) * np.exp(-fase)
  return re

resis = calcularresistencia(fase)
x = resis * np.cos(fase)
y = resis * np.sin(fase)

#frecuencia angular grande
plt.plot(x, y)
plt.grid()
plt.axhline(0, color='yellow')
plt.axvline(0, color='yellow')
plt.xlim(-0.8, 0.8)
plt.ylim(-0.8, 0.8)
plt.show()
```
